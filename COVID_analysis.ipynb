```python
df = df[['location', 'date', 'total_cases', 'total_deaths', 'total_vaccinations',
         'new_cases', 'new_deaths', 'population', 'gdp_per_capita', 'hospital_beds_per_thousand']]
df['date'] = pd.to_datetime(df['date'])
df = df.fillna(0)
```

```python
global_details = {
    "Total Cases":df["total_cases"].sum(),
    "Total Deaths":df["total_deaths"].sum(),
     "Total Vaccinations":df["total_vaccinations"].sum()
}
print(global_details)
```

```python
# Task 1: Total cases, deaths, and vaccinations globally
global_cases = df['total_cases'].max()
global_deaths = df['total_deaths'].max()
global_vaccinations = df['total_vaccinations'].max()
```

```python
print(f"Total Global Cases: {global_cases}")
print(f"Total Global Deaths: {global_deaths}")
print(f"Total Global Vaccinations: {global_vaccinations}")
```

```python
# Task 2: Countries with the highest and lowest total cases and deaths
latest_data = df[df['date'] == df['date'].max()]
highest_cases = latest_data.loc[latest_data['total_cases'].idxmax()]
lowest_cases = latest_data.loc[latest_data['total_cases'].idxmin()]
highest_deaths = latest_data.loc[latest_data['total_deaths'].idxmax()]
lowest_deaths = latest_data.loc[latest_data['total_deaths'].idxmin()]
```

```python
print("\nCountry with Highest Cases:")
print(highest_cases[['location', 'total_cases']])
print("\nCountry with Lowest Cases:")
print(lowest_cases[['location', 'total_cases']])
print("\nCountry with Highest Deaths:")
print(highest_deaths[['location', 'total_deaths']])
print("\nCountry with Lowest Deaths:")
print(lowest_deaths[['location', 'total_deaths']])
```

```python
# Task 3: Global Case Fatality Rate (CFR)
cfr = (global_deaths / global_cases) * 100
print(f"\nGlobal Case Fatality Rate (CFR): {cfr:.2f}%")
```

```python
# Task 4: New cases and deaths trends over time
global_daily_data = df.groupby('date').sum()

plt.figure(figsize=(12, 6))
plt.plot(global_daily_data.index, global_daily_data['new_cases'], label='New Cases', color='blue')
plt.plot(global_daily_data.index, global_daily_data['new_deaths'], label='New Deaths', color='red')
plt.title("Global New Cases and Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.grid()
plt.show()
```

```python
#CALCULATING CASES:
df['total_cases_per_million'] = (df['total_cases'] / df['population']) * 1e6

#CALCULATING DEATHS:
df['total_deaths_per_million'] = (df['total_deaths'] / df['population']) * 1e6

top_cases = df[['location', 'total_cases_per_million']].sort_values(by='total_cases_per_million', ascending=False)
print(top_cases.head(10))

top_deaths = df[['location', 'total_deaths_per_million']].sort_values(by='total_deaths_per_million', ascending=False)
print(top_deaths.head(10))
```

```python
df['CFR'] = (df['total_deaths'] / df['total_cases']) * 100

plt.figure(figsize=(12, 6))
sns.scatterplot(x='hospital_beds_per_thousand', y='CFR', data=df)
plt.title('Relation Between Hospital Beds per Thousand and Case Fatality Rate (CFR)')
plt.xlabel('Hospital Beds per Thousand')
plt.ylabel('Case Fatality Rate')
plt.tight_layout()
plt.show()
```

```python
#CALCULATING THE POSITIVE TEST RATE OF COVID-19:
df['positive_rate'] = df['new_cases'] / df['new_tests']

#CORRELATION BETWEEN POSITIVE TEST RATES ND NEW CASES:
correlation_positive_rate_cases = df['positive_rate'].corr(df['new_cases'])

print(f"Correlation between Positive Test Rate and New Cases: {correlation_positive_rate_cases}")

#SCATTER PLOT:
plt.figure(figsize=(12, 6))
plt.scatter(df['positive_rate'], df['new_cases'], alpha=0.5)
plt.title('Positive Test Rate vs New Cases')
plt.xlabel('Positive Test Rate')
plt.ylabel('New Cases')
plt.tight_layout()
plt.show()
```

```python
#CALCULATING POSITIVITY RATES:
df['positive_rate'] = df['new_cases'] / df['new_tests']

#GROUPIING BY COUNTRY AND CALCULATING THE MEAN POSITIVITY RATE OVER TIME:
positive_rate_by_country = df.groupby(['location', 'date'])['positive_rate'].mean().unstack('location')

#REPLACING OF NAME SOF FEW COUNTRIES FOR TESTING FUTHER:
countries_to_plot = ['United States', 'India', 'Brazil']  # Replace with countries that are in the dataset

#PLOTTING:
positive_rate_by_country[countries_to_plot].plot(figsize=(12, 6))
plt.title('Test Positivity Rates by Country Over Time')
plt.xlabel('Date')
plt.ylabel('Test Positivity Rate')
plt.tight_layout()
plt.show()
```

```python
#CALCULATING THE TEST PER RATE RATIO:
df['tests_per_case'] = df['new_tests'] / df['new_cases']

#SORTING BY TEST PER CASE RATIO AND DISPLAY THE HIGHEST AND THE LOWEST VALUES;
highest_tests_per_case = df[['location', 'tests_per_case']].dropna().sort_values('tests_per_case', ascending=False).head(10)
lowest_tests_per_case = df[['location', 'tests_per_case']].dropna().sort_values('tests_per_case', ascending=True).head(10)
print("Countries with the Highest Tests per Case Ratio:")
print(highest_tests_per_case)
print("\nCountries with the Lowest Tests per Case Ratio:")
print(lowest_tests_per_case)
```

```python
top_affected = df.groupby('location')[['total_cases', 'total_deaths']].max()
top_affected_by_cases = top_affected.sort_values('total_cases', ascending=False).head(10)
top_affected_by_deaths = top_affected.sort_values('total_deaths', ascending=False).head(10)
print("Top 10 by Cases:\n", top_affected_by_cases)
print("Top 10 by Deaths:\n", top_affected_by_deaths)
```

```python
#CHECKING THE RESPECTIVE COLUMN:
if 'total_cases' in df.columns and 'total_deaths' in df.columns:
    # Global CFR (total deaths / total cases)
    global_cases = df['total_cases'].sum()
    global_deaths = df['total_deaths'].sum()


    if global_cases > 0:
        global_cfr = global_deaths / global_cases
    else:
        global_cfr = 0

    print(f"Global Case Fatality Rate (CFR): {global_cfr:.4f}")

    #CALCULATING CFER BY COUNTRY:
    cfr_by_country = df.groupby('location').agg({
        'total_cases': 'sum',
        'total_deaths': 'sum'
    })

    cfr_by_country['CFR'] = cfr_by_country['total_deaths'] / cfr_by_country['total_cases']

    #SORTING COUNTRIES BY CFR:
    cfr_by_country_sorted = cfr_by_country.sort_values(by='CFR', ascending=False)

    #DISPLAY OF 10 COUNTRIES THROUGH CFR:
    print("\nTop 10 countries by Case Fatality Rate (CFR):")
    print(cfr_by_country_sorted[['CFR']].head(10))

else:
    print("Columns 'total_cases' or 'total_deaths' are not present in the dataset.")
```

```python
# Calculating the slope of new cases for each country (e.g., linear regression or simple difference):
df['new_cases_slope'] = df.groupby('location')['new_cases'].diff()

#Threshold for a "flattened curve":
threshold = 5

#Identify countries with a "flattened curve" (low slope of new cases):
flattened_countries = df.groupby('location')['new_cases_slope'].mean()
flattened_countries = flattened_countries[flattened_countries.abs() < threshold]  # Flattened curve: small slope

#Countries with ongoing waves (high slope of new cases):
ongoing_countries = df.groupby('location')['new_cases_slope'].mean()
ongoing_countries = ongoing_countries[ongoing_countries.abs() >= threshold]

#Output the countries with flattened and ongoing curves:
print("Flattened Curve Countries:", flattened_countries.index.tolist())
print("Ongoing Curve Countries:", ongoing_countries.index.tolist())
```
